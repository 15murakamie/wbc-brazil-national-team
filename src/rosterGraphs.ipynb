#### ParticipaÃ§Ãµes por jogador e treinador

```python
# ParticipaÃ§Ã£o por jogador
player_participation = df_events_player.groupby("nome").size().reset_index(name="participacoes")
# ParticipaÃ§Ã£o por treinador
coach_participation = df_events_coach.groupby("nome").size().reset_index(name="participacoes")


# Fazer o merge (junÃ§Ã£o) dos dois DataFrames pela coluna 'nome'
merged_participation = pd.merge(
    player_participation,
    coach_participation,
    on="nome",
    how="outer",  # Garante que quem participou sÃ³ como jogador ou sÃ³ como treinador tambÃ©m apareÃ§a
    suffixes=("_player", "_coach")
)

# Preencher valores ausentes com 0
merged_participation["participacoes_player"] = merged_participation["participacoes_player"].fillna(0).astype(int)
merged_participation["participacoes_coach"] = merged_participation["participacoes_coach"].fillna(0).astype(int)

# Criar coluna de total de participaÃ§Ãµes
merged_participation["participacoes_total"] = (
    merged_participation["participacoes_player"] + merged_participation["participacoes_coach"]
)

# Ordenar pela soma total de participaÃ§Ãµes (ordem decrescente)
merged_participation = merged_participation.sort_values(by=["participacoes_total","participacoes_player","participacoes_coach","nome"], ascending=False).reset_index(drop=True).head(11)


merged_participation
```

```python
import plotly.graph_objects as go

# Para inverter, criamos a lista invertida de nomes
nomes_invertidos = merged_participation["nome"][::-1]

fig = go.Figure()

fig.add_trace(go.Bar(
    y=merged_participation["nome"],
    x=merged_participation["participacoes_player"],
    name="Jogador",
    marker_color="royalblue",
    orientation='h',
    text=merged_participation["participacoes_player"],
    textposition='inside',
    insidetextanchor='middle',
))

fig.add_trace(go.Bar(
    y=merged_participation["nome"],
    x=merged_participation["participacoes_coach"],
    name="Treinador",
    marker_color="orange",
    orientation='h',
    text=merged_participation["participacoes_coach"],
    textposition='inside',
    insidetextanchor='middle',
))

fig.update_layout(
    barmode="stack",
    title="ParticipaÃ§Ãµes no WBC",
    xaxis_title="NÃºmero de ParticipaÃ§Ãµes",
    yaxis_title="Nome do Jogagor/Treinador",
    width=800,
    height=500,
    margin=dict(l=150, r=50, t=100, b=50),
    yaxis=dict(
        categoryorder='array',
        categoryarray=nomes_invertidos,
        automargin=True,
        tickfont=dict(size=12),
    )
)

fig.show()
```

#### AnÃ¡lise da Idade dos Jogadores

```python
# Criando faixas etÃ¡rias
bins = [0, 19, 24, 29, 34, 100]
labels = ["<20", "20-24", "25-29", "30-34", "35+"]

# Convertendo para os formatos desejados
df_events_player["idade"] = pd.to_numeric(df_events_player["idade"], errors="coerce")
df_events_player["anoTorneio_str"] = df_events_player["anoTorneio"].astype(str)


# Definindo faixa etÃ¡ria
df_events_player["faixa_etaria"] = pd.cut(df_events_player["idade"], bins=bins, labels=labels, right=True)

# Agrupar: contar jogadores por ano e faixa etÃ¡ria
df_grouped = df_events_player.groupby(["anoTorneio_str", "faixa_etaria"]).size().reset_index(name="quantidade")
```

```python
df_events_player.head()
```

```python
# FunÃ§Ã£o para extrair o nome do jogador mais jovem e mais velho
def get_jogador_extremo(grupo, mais_velho=True):
    if mais_velho:
        return grupo.loc[grupo['idade'].idxmax(), 'nome']
    else:
        return grupo.loc[grupo['idade'].idxmin(), 'nome']

# Agrupando por ano
resumo_idades = df_events_player.groupby('anoTorneio').agg(
    idadeMin=('idade', 'min'),
    idadeMax=('idade', 'max'),
    idadeMedia=('idade', 'mean')
).reset_index()

# Adicionando nomes dos jogadores mais jovem e mais velho
resumo_idades['jogadorMaisJovem'] = df_events_player.groupby('anoTorneio').apply(
    lambda x: get_jogador_extremo(x, mais_velho=False)
).values

resumo_idades['jogadorMaisVelho'] = df_events_player.groupby('anoTorneio').apply(
    lambda x: get_jogador_extremo(x, mais_velho=True)
).values

# Opcional: arredondar a mÃ©dia para uma casa decimal
resumo_idades['idadeMedia'] = resumo_idades['idadeMedia'].round(1)

# Visualizar
resumo_idades
```

#### Ligas onde jÃ¡ Atuaram

```python
import plotly.express as px
import plotly.graph_objects as go

# Colunas de ligas
ligas = ['isMLB', 'isNPB', 'isMEX', 'isVEN', 'isMiLB', 'isJPIND']

# De-para das ligas
nomes_ligas = {
    'isMLB': 'MLB',
    'isNPB': 'NPB (Liga Japonesa)',
    'isMEX': 'Liga Mexicana',
    'isVEN': 'Liga Venezuelana',
    'isMiLB': 'MiLB - Minor League',
    'isJPIND': 'Ligas Independentes Japonesas'
}

# Soma total por liga
liga_totais = df_player_list[ligas].sum().reset_index()
liga_totais.columns = ['Liga', 'Quantidade']
liga_totais['Liga'] = liga_totais['Liga'].replace(nomes_ligas)

# GrÃ¡fico de pizza
fig_pie = px.pie(
    liga_totais,
    names='Liga',
    values='Quantidade',
    title='DistribuiÃ§Ã£o de Jogadores por Liga Profissional',
    color_discrete_sequence=px.colors.qualitative.Set3,
    width=600,
    height=500,
)

fig_pie.update_layout(
    title_y=0.85  # centraliza o tÃ­tulo (0 = esquerda, 1 = direita)
)

# Texto dentro dos setores
fig_pie.update_traces(textposition='inside', textinfo='value')

# Adiciona legenda explicativa no rodapÃ©
fig_pie.add_annotation(
    text="Dados acumulados atÃ© 05/2025, independentemente do ano do torneio.",
    xref="paper", yref="paper",
    x=0.5, y=-0.05,  # posiÃ§Ã£o abaixo do grÃ¡fico
    showarrow=False,
    font=dict(size=10),
    align="center"
)

fig_pie.show()
```

```python
import plotly.express as px

# Prepara os dados
df_player_list[ligas] = df_player_list[ligas].fillna(0).astype(int)

# Extrai paÃ­s do local de nascimento
df_player_list['pais'] = df_player_list['localNascimento'].str.extract(r',\s*(\w+)$', expand=False)

# Agrupa por paÃ­s
df_pais_ligas = df_player_list.groupby('pais')[ligas].sum().reset_index()

# Remove paÃ­ses sem participaÃ§Ã£o
df_pais_ligas = df_pais_ligas[df_pais_ligas[ligas].sum(axis=1) > 0]

# Ordena paÃ­ses por total
df_pais_ligas['total'] = df_pais_ligas[ligas].sum(axis=1)
df_pais_ligas = df_pais_ligas.sort_values(by='total', ascending=False).drop(columns='total')

# Renomeia coluna
df_pais_ligas = df_pais_ligas.rename(columns={'pais': 'PaÃ­s de Origem'})

# Nomes mais amigÃ¡veis para ligas
label_map = {
    'isMLB': 'Americana (MLB)', 'isNPB': 'Japonesa (NPB)', 'isKBO': 'Coreana',
    'isMEX': 'Mexicana', 'isTWN': 'Taiwanesa',
    'isVEN': 'Venezuelana', 'isMiLB': 'Minor (MiLB)',
    'isJPIND': 'Indep. Japonesa'
}

# Formato longo
df_melted = df_pais_ligas.melt(id_vars='PaÃ­s de Origem', value_vars=ligas, var_name='Liga', value_name='Quantidade')
df_melted['Liga'] = df_melted['Liga'].replace(label_map)

# Ordena as ligas pelo total geral (mais usadas no topo)
liga_order = df_melted.groupby('Liga')['Quantidade'].sum().sort_values(ascending=False).index.tolist()
df_melted['Liga'] = pd.Categorical(df_melted['Liga'], categories=liga_order, ordered=True)

# Heatmap com melhorias
fig_heatmap = px.density_heatmap(
    df_melted,
    x="PaÃ­s de Origem",
    y="Liga",
    z="Quantidade",
    color_continuous_scale="YlGnBu",
    title="NÃºmero de Jogadores por Liga e PaÃ­s de Nascimento (atÃ© 2025)",
    text_auto=True,
    width=800,
    height=520
)

# Layout aprimorado
fig_heatmap.update_layout(
    coloraxis_colorbar_title="NÂ° de Jogadores",
    title_x=0.5,
    margin=dict(l=40, r=40, t=60, b=80),
)

# Adiciona legenda explicativa (rodapÃ©)
fig_heatmap.add_annotation(
    text="Dados acumulados atÃ© maio de 2025 â€” independentemente do ano da convocaÃ§Ã£o.",
    xref="paper", yref="paper",
    x=0.5, y=-0.22,
    showarrow=False,
    font=dict(size=12),
    align="center"
)

fig_heatmap.show()
```
